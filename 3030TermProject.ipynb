```python
# #2==========DEFINE DATA CLEANING FUNCTIONS=======
def dateCleanFB(rowUnix,rowDate):
    if bool(re.match('.*/2',rowUnix))==True:
        return pd.to_datetime(rowDate, format='%m/%d/%Y')
    else:
        return pd.to_datetime(rowDate, format='%d/%m/%Y')

def dateCleanTweet(rowDate):
    return pd.to_datetime(rowDate, format='%d/%m/%Y')

def dateCleanWebI(rowDate):
    return pd.to_datetime(rowDate, format='%m/%d/%Y')
```
